```python
import os
import cv2
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from scipy.io import loadmat
from scipy.ndimage import gaussian_filter
from PIL import Image
```

```python
class CrowdDataset(Dataset):
    def __init__(self, img_folder, gt_folder, transform=None, target_size=(512, 512)):
        self.img_folder = img_folder
        self.gt_folder = gt_folder
        self.transform = transform
        self.target_size = target_size
        self.img_files = [f for f in os.listdir(img_folder) if f.endswith('.jpg')]

    def __len__(self):
        return len(self.img_files)

    def __getitem__(self, idx):
        img_name = self.img_files[idx]
        img_path = os.path.join(self.img_folder, img_name)
        gt_path = os.path.join(self.gt_folder, "GT_" + img_name.replace(".jpg", ".mat"))

        # Load image
        img = Image.open(img_path).convert('RGB')

        # Load ground-truth points
        mat = loadmat(gt_path)
        points = mat["image_info"][0][0][0][0][0]

        # Create density map
        density_map = self.generate_density_map(img, points)

        # Resize image and density map
        img = img.resize(self.target_size)
        density_map = cv2.resize(density_map, self.target_size)
        density_map = density_map * ((density_map.sum() + 1e-5) / (density_map.sum() + 1e-5))

        if self.transform:
            img = self.transform(img)

        density_map = torch.from_numpy(density_map).unsqueeze(0).float()
        return img, density_map

    def generate_density_map(self, img, points):
        w, h = img.size
        density = np.zeros((h, w), dtype=np.float32)
        for x, y in points:
            if int(y) < h and int(x) < w:
                density[int(y), int(x)] = 1
        density = gaussian_filter(density, sigma=15)
        return density
```

```python
import torch.nn.functional as F

class CSRNet(nn.Module):
    def __init__(self):
        super(CSRNet, self).__init__()
        self.frontend = nn.Sequential(
            nn.Conv2d(3, 64, 3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, 3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2),
            nn.Conv2d(64, 128, 3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, 128, 3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2)
        )

        self.backend = nn.Sequential(
            nn.Conv2d(128, 512, 3, padding=2, dilation=2),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 512, 3, padding=2, dilation=2),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 512, 3, padding=2, dilation=2),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 256, 3, padding=2, dilation=2),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 128, 3, padding=2, dilation=2),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, 1, 1)
        )

    def forward(self, x):
        x = self.frontend(x)
        x = self.backend(x)
        x = F.interpolate(x, size=(512, 512), mode='bilinear', align_corners=False)
        return x
```

```python
import os
import numpy as np
import h5py
from scipy.io import loadmat
from scipy.ndimage import gaussian_filter
from PIL import Image

# Paths
image_folder = "dataset/ShanghaiTech/part_A/train_data/images"
mat_folder = "dataset/ShanghaiTech/part_A/train_data/ground-truth"
output_folder = "dataset/ShanghaiTech/part_A/train_data/ground-truth_npy"

os.makedirs(output_folder, exist_ok=True)

def generate_density_map(image_path, mat_path):
    """Generate density map from .mat annotation file."""
    # Load image to get shape
    img = Image.open(image_path)
    width, height = img.size

    # Load .mat file
    mat = loadmat(mat_path)
    if "image_info" in mat:
        points = mat["image_info"][0,0][0,0][0]  # For ShanghaiTech Part A/B
    else:
        points = mat["annPoints"]  # For other crowd datasets

    # Create empty map
    density = np.zeros((height, width), dtype=np.float32)

    # Add Gaussian for each annotation point
    for point in points:
        x, y = min(int(point[0]), width-1), min(int(point[1]), height-1)
        density[y, x] += 1

    density = gaussian_filter(density, sigma=15)
    return density

# Convert all files
for img_name in os.listdir(image_folder):
    if not img_name.endswith(".jpg"):
        continue
    img_path = os.path.join(image_folder, img_name)
    mat_path = os.path.join(mat_folder, f"GT_{img_name.split('.')[0]}.mat")

    if os.path.exists(mat_path):
        density = generate_density_map(img_path, mat_path)
        np.save(os.path.join(output_folder, img_name.replace(".jpg", ".npy")), density)

print(" Density maps generated and saved in:", output_folder)
print("Total generated:", len(os.listdir(output_folder)))
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import models

# CSRNet-like model
class CSRNet(nn.Module):
    def __init__(self):
        super(CSRNet, self).__init__()
        vgg = models.vgg16(pretrained=True)
        features = list(vgg.features.children())[:23]  # first 10 conv layers of VGG16
        self.frontend = nn.Sequential(*features)

        self.backend = nn.Sequential(
            nn.Conv2d(512, 512, 3, padding=2, dilation=2),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 512, 3, padding=2, dilation=2),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 512, 3, padding=2, dilation=2),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 256, 3, padding=2, dilation=2),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 128, 3, padding=2, dilation=2),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, 64, 1),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 1, 1)
        )

    def forward(self, x):
        x = self.frontend(x)
        x = self.backend(x)
        return x

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = CSRNet().to(device)

print(" Model loaded on device:", device)
```

```python
import os
import numpy as np
from PIL import Image
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import models, transforms
from torch.optim import Adam
from torch.optim.lr_scheduler import StepLR
import matplotlib.pyplot as plt

# ============================================================
# 1 Dataset Class
# ============================================================
class CrowdDataset(Dataset):
    def __init__(self, img_folder, density_folder, transform=None, target_size=(256, 256)):
        self.img_folder = img_folder
        self.density_folder = density_folder
        self.transform = transform
        self.target_size = target_size
        self.images = [f for f in os.listdir(img_folder) if f.endswith('.jpg')]

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img_name = self.images[idx]
        img_path = os.path.join(self.img_folder, img_name)
        density_path = os.path.join(self.density_folder, img_name.replace('.jpg', '.npy'))

        # Load image
        image = Image.open(img_path).convert('RGB')
        if self.transform:
            image = self.transform(image)

        # Load and resize density map
        density = np.load(density_path)
        density = torch.from_numpy(density).unsqueeze(0).float()  # [1, H, W]
        original_sum = density.sum()

        # Resize using bilinear interpolation
        density = F.interpolate(
            density.unsqueeze(0),
            size=self.target_size,
            mode='bilinear',
            align_corners=False
        ).squeeze(0)

        #  Preserve total count after resize
        if density.sum() > 0:
            density *= (original_sum / density.sum())

        return image, density

# ============================================================
# 2 Transformations
# ============================================================
transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225]),
])

# ============================================================
# 3 Dataset & DataLoader
# ============================================================
train_dataset = CrowdDataset(
    "dataset/ShanghaiTech/part_A/train_data/images",
    "dataset/ShanghaiTech/part_A/train_data/ground-truth_npy",
    transform=transform,
    target_size=(256, 256)
)

train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True, num_workers=0)
print(" Total training samples:", len(train_dataset))

# ============================================================
# 4 Model Definition
# ============================================================
class CrowdCounter(nn.Module):
    def __init__(self):
        super(CrowdCounter, self).__init__()
        vgg = models.vgg16(weights=models.VGG16_Weights.IMAGENET1K_V1)
        self.features = nn.Sequential(*list(vgg.features.children())[:-2])
        self.regressor = nn.Sequential(
            nn.Conv2d(512, 256, 3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, 1, 1)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.regressor(x)
        return x

# ============================================================
# 5 Device Setup
# ============================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = CrowdCounter().to(device)
print(f" Model loaded on device: {device}")

# ============================================================
# 6 Training Setup
# ============================================================
criterion = nn.MSELoss()
optimizer = Adam(model.parameters(), lr=1e-5, weight_decay=1e-4)
scheduler = StepLR(optimizer, step_size=15, gamma=0.5)
max_epochs = 50

train_losses = []

# ============================================================
# 7 Training Loop
# ============================================================
for epoch in range(max_epochs):
    model.train()
    running_loss = 0.0

    for images, density_maps in train_loader:
        images = images.to(device)
        density_maps = density_maps.to(device)

        optimizer.zero_grad()
        outputs = model(images)

        # Match density map size to model output
        density_maps_resized = F.interpolate(
            density_maps, size=outputs.shape[2:], mode='bilinear', align_corners=False
        )

        # Keep total count consistent
        density_maps_resized *= (density_maps.sum(dim=[1,2,3], keepdim=True) /
                                 density_maps_resized.sum(dim=[1,2,3], keepdim=True))

        loss = criterion(outputs, density_maps_resized)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

    avg_loss = running_loss / len(train_loader)
    train_losses.append(avg_loss)
    print(f"Epoch [{epoch+1}/{max_epochs}] Loss: {avg_loss:.6f}")

    scheduler.step()

print(" Training complete!")

# ============================================================
# 8 Save Model
# ============================================================
torch.save(model.state_dict(), "crowd_counter_final.pth")
print(" Model saved as 'crowd_counter_final.pth'")

# ============================================================
# 9 Plot Training Loss
# ============================================================
plt.plot(train_losses)
plt.title("Training Loss over Epochs")
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.show()

# ============================================================
#  Visualize Density Map Prediction
# ============================================================
model.eval()
with torch.no_grad():
    img, gt_density = train_dataset[0]
    img_input = img.unsqueeze(0).to(device)
    pred_density = model(img_input).squeeze().cpu().numpy()
    gt_density = gt_density.squeeze().numpy()

plt.figure(figsize=(12,5))
plt.subplot(1,3,1)
plt.title("Input Image")
plt.imshow(np.transpose(img.numpy(), (1,2,0)))

plt.subplot(1,3,2)
plt.title(f"GT Density (sum={gt_density.sum():.1f})")
plt.imshow(gt_density, cmap='jet')

plt.subplot(1,3,3)
plt.title(f"Pred Density (sum={pred_density.sum():.1f})")
plt.imshow(pred_density, cmap='jet')

plt.show()


```

```python
def unnormalize(tensor):
    """Undo ImageNet normalization for visualization."""
    mean = torch.tensor([0.485, 0.456, 0.406]).view(3,1,1)
    std = torch.tensor([0.229, 0.224, 0.225]).view(3,1,1)
    tensor = tensor * std + mean
    tensor = torch.clamp(tensor, 0, 1)
    return tensor
```

```python
# ================================================================
#  Real-Time Crowd Counting (Accurate + Clean Output)
# ================================================================
import os
import cv2
import torch
import numpy as np
import torch.nn as nn
import torch.nn.functional as F
from torchvision import models, transforms
from collections import deque
import winsound  # for alert sound (Windows)

# ================================================================
# 1 Model Definition
# ================================================================
class CrowdCounter(nn.Module):
    def __init__(self):
        super(CrowdCounter, self).__init__()
        vgg = models.vgg16(weights=models.VGG16_Weights.IMAGENET1K_V1)
        self.features = nn.Sequential(*list(vgg.features.children())[:-2])
        self.regressor = nn.Sequential(
            nn.Conv2d(512, 256, 3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, 1, 1)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.regressor(x)
        return x

# ================================================================
# 2 Load Trained Model
# ================================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = CrowdCounter().to(device)

model_path = "crowd_counter_final.pth"
if not os.path.exists(model_path):
    raise FileNotFoundError(f" Model file not found: {model_path}")

state_dict = torch.load(model_path, map_location=device)
model.load_state_dict(state_dict)
model.eval()
print(" Model loaded successfully for real-time inference!")

# ================================================================
# 3 Transform (MUST match training)
# ================================================================
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Resize((512, 512)),  # same as training
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

# ================================================================
# 4 Real-Time Video Stream
# ================================================================
CROWD_THRESHOLD = 50       # adjust as needed
CALIBRATION_FACTOR = 0.95  # fine-tune for your dataset
SAVE_VIDEO = False         # no saving to keep it lightweight

cap = cv2.VideoCapture(0)  # 0 = webcam; can replace with video file path
if not cap.isOpened():
    raise RuntimeError(" Could not open video stream.")

fps = int(cap.get(cv2.CAP_PROP_FPS)) or 25
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

print(" Starting real-time crowd counting... Press 'q' to quit.")

# ================================================================
# 5 Smooth Counting (Moving Average)
# ================================================================
recent_counts = deque(maxlen=5)

# ================================================================
# 6 Real-Time Loop
# ================================================================
while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Convert frame to RGB and preprocess
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    img_tensor = transform(rgb_frame).unsqueeze(0).to(device)

    # Predict density map
    with torch.no_grad():
        density_map = model(img_tensor)
        density_map = F.interpolate(density_map, size=(frame.shape[0], frame.shape[1]),
                                    mode='bilinear', align_corners=False)
        density = density_map.squeeze().cpu().numpy()
        crowd_count = np.sum(density) * CALIBRATION_FACTOR

    # Apply moving average for smoother count
    recent_counts.append(crowd_count)
    smoothed_count = np.mean(recent_counts)

    # Display clean count only
    text = f"Crowd Count: {int(smoothed_count)}"
    color = (0, 255, 0)

    if smoothed_count > CROWD_THRESHOLD:
        color = (0, 0, 255)
        text += "  Overcrowded!"
        winsound.Beep(1000, 300)  # alert sound

    cv2.putText(frame, text, (20, 50), cv2.FONT_HERSHEY_SIMPLEX, 1.2, color, 3)
    cv2.imshow("Real-Time Crowd Counting", frame)

    # Quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

print(" Real-time crowd counting stopped successfully!")
```

```python
"""
webcam_person_counter_save_video.py

- Uses OpenCV HOG person detector.
- Applies Non-Max Suppression (NMS).
- Tracks people with a simple centroid tracker to maintain stable integer counts.
- Saves output video (annotated) to crowd_count_output.mp4.
- Press 'q' to quit.

Works offline (no extra model downloads). If you want even better accuracy, replace the detect_persons(...) function
with a DNN-based detector (YOLO / MobileNet-SSD).
"""

import cv2
import numpy as np
import time
from collections import OrderedDict
from scipy.spatial import distance as dist
import os

# ----------------------------
# CentroidTracker (simple)
# ----------------------------
class CentroidTracker:
    def __init__(self, maxDisappeared=30, maxDistance=50):
        # next unique object ID
        self.nextObjectID = 0
        # objectID -> centroid
        self.objects = OrderedDict()
        # objectID -> number of consecutive frames it has disappeared
        self.disappeared = OrderedDict()
        self.maxDisappeared = maxDisappeared
        self.maxDistance = maxDistance

    def register(self, centroid, bbox):
        self.objects[self.nextObjectID] = (centroid, bbox)
        self.disappeared[self.nextObjectID] = 0
        self.nextObjectID += 1

    def deregister(self, objectID):
        del self.objects[objectID]
        del self.disappeared[objectID]

    def update(self, rects):
        """
        rects: list of bounding boxes (x, y, w, h)
        returns: OrderedDict of objectID -> (centroid, bbox)
        """
        # if no detections, mark all existing as disappeared
        if len(rects) == 0:
            for objectID in list(self.disappeared.keys()):
                self.disappeared[objectID] += 1
                if self.disappeared[objectID] > self.maxDisappeared:
                    self.deregister(objectID)
            return self.objects

        # compute input centroids
        inputCentroids = []
        inputBBoxes = []
        for (x, y, w, h) in rects:
            cX = int(x + w / 2.0)
            cY = int(y + h / 2.0)
            inputCentroids.append((cX, cY))
            inputBBoxes.append((x, y, w, h))

        # if no existing objects, register all
        if len(self.objects) == 0:
            for i in range(len(inputCentroids)):
                self.register(inputCentroids[i], inputBBoxes[i])
        else:
            # build matrix of distances between each pair of existing object centroids and new centroids
            objectIDs = list(self.objects.keys())
            objectCentroids = [self.objects[oid][0] for oid in objectIDs]

            D = dist.cdist(np.array(objectCentroids), np.array(inputCentroids))
            # for each existing object, find the nearest new centroid
            rows = D.min(axis=1).argsort()
            cols = D.argmin(axis=1)[rows]

            usedRows = set()
            usedCols = set()

            for (row, col) in zip(rows, cols):
                if row in usedRows or col in usedCols:
                    continue
                if D[row, col] > self.maxDistance:
                    continue

                objectID = objectIDs[row]
                self.objects[objectID] = (inputCentroids[col], inputBBoxes[col])
                self.disappeared[objectID] = 0

                usedRows.add(row)
                usedCols.add(col)

            # check unmatched existing object rows -> mark disappeared
            unusedRows = set(range(0, D.shape[0])).difference(usedRows)
            for row in unusedRows:
                objectID = objectIDs[row]
                self.disappeared[objectID] += 1
                if self.disappeared[objectID] > self.maxDisappeared:
                    self.deregister(objectID)

            # check unmatched new detection cols -> register them
            unusedCols = set(range(0, D.shape[1])).difference(usedCols)
            for col in unusedCols:
                self.register(inputCentroids[col], inputBBoxes[col])

        return self.objects

# ----------------------------
# Helper: non-max suppression for rectangles
# ----------------------------
def non_max_suppression_fast(boxes, overlapThresh):
    # boxes: list of (x, y, w, h)
    if len(boxes) == 0:
        return []
    # convert to array [x1,y1,x2,y2]
    rects = np.array([[x, y, x + w, y + h] for (x, y, w, h) in boxes], dtype=float)
    pick = []
    x1 = rects[:,0]
    y1 = rects[:,1]
    x2 = rects[:,2]
    y2 = rects[:,3]
    area = (x2 - x1 + 1) * (y2 - y1 + 1)
    idxs = np.argsort(y2)
    while len(idxs) > 0:
        last = idxs[-1]
        pick.append(last)
        xx1 = np.maximum(x1[last], x1[idxs[:-1]])
        yy1 = np.maximum(y1[last], y1[idxs[:-1]])
        xx2 = np.minimum(x2[last], x2[idxs[:-1]])
        yy2 = np.minimum(y2[last], y2[idxs[:-1]])
        w = np.maximum(0, xx2 - xx1 + 1)
        h = np.maximum(0, yy2 - yy1 + 1)
        overlap = (w * h) / area[idxs[:-1]]
        idxs = np.delete(idxs, np.concatenate(([len(idxs)-1], np.where(overlap > overlapThresh)[0])))
    # return selected boxes in original (x,y,w,h) format
    return [tuple(map(int, (x1[i], y1[i], x2[i]-x1[i], y2[i]-y1[i]))) for i in pick]

# ----------------------------
# Person detection function (HOG)
# ----------------------------
def detect_persons(frame, winStride=(8,8), padding=(8,8), scale=1.05, nms_thresh=0.3):
    """
    return: list of bounding boxes (x,y,w,h)
    """
    # HOG returns (x, y, w, h) and weights
    rects, weights = hog.detectMultiScale(frame, winStride=winStride, padding=padding, scale=scale)
    rects = list(rects)
    # apply non-max suppression for stability
    pick = non_max_suppression_fast(rects, overlapThresh=nms_thresh)
    return pick

# ----------------------------
# Initialize HOG detector
# ----------------------------
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())

# ----------------------------
# Create output directory
# ----------------------------
OUTPUT_FILE = "crowd_count_output.mp4"
fourcc = cv2.VideoWriter_fourcc(*'mp4v')

# ----------------------------
# Start capture
# ----------------------------
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise RuntimeError("Could not open webcam")

# read a frame to get size
ret, frame = cap.read()
if not ret:
    raise RuntimeError("Could not read webcam frame")
frame = cv2.resize(frame, (640, 480))
(h, w) = frame.shape[:2]

out = cv2.VideoWriter(OUTPUT_FILE, fourcc, 20.0, (w, h))
print("Saving annotated video to:", OUTPUT_FILE)

ct = CentroidTracker(maxDisappeared=25, maxDistance=60)

# parameters (tweak if needed)
WIN_STRIDE = (8, 8)
PADDING = (8, 8)
SCALE = 1.05
NMS_THRESH = 0.3
CROWD_ALERT_THRESHOLD = 3  # example threshold to trigger alert

# To avoid repeated beeps flooding, remember last beep time
last_beep = 0
BEEP_INTERVAL = 2.0  # seconds

print("Press 'q' to quit.")
while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.resize(frame, (w, h))

    # detect persons (on color frame)
    boxes = detect_persons(frame, winStride=WIN_STRIDE, padding=PADDING, scale=SCALE, nms_thresh=NMS_THRESH)

    # update tracker
    objects = ct.update(boxes)

    # draw boxes and IDs
    for objectID, (centroid, bbox) in objects.items():
        (x, y, bw, bh) = bbox
        # ensure ints
        x, y, bw, bh = int(x), int(y), int(bw), int(bh)
        cv2.rectangle(frame, (x, y), (x + bw, y + bh), (0, 255, 0), 2)
        text = f"ID {objectID}"
        cv2.putText(frame, text, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,255,0), 2)
        cv2.circle(frame, centroid, 3, (0, 0, 255), -1)

    # current count = number of tracked objects
    crowd_count = len(objects)

    # display count on frame
    color = (0, 255, 0)
    alert_text = ""
    if crowd_count == 1:
        color = (0, 255, 0)
        alert_text = "One person detected"
    elif crowd_count > CROWD_ALERT_THRESHOLD:
        color = (0, 0, 255)
        alert_text = "Overcrowded!"
        # beep occasionally
        if time.time() - last_beep > BEEP_INTERVAL:
            try:
                import winsound
                winsound.Beep(1000, 200)
            except Exception:
                pass
            last_beep = time.time()

    cv2.putText(frame, f"Crowd Count: {crowd_count}", (10, 25), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)
    if alert_text:
        cv2.putText(frame, alert_text, (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)

    # write frame to output
    out.write(frame)

    # show frame
    cv2.imshow("Person Counter (press q to quit)", frame)

    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'):
        break

# cleanup
cap.release()
out.release()
cv2.destroyAllWindows()
print("Saved video:", os.path.abspath(OUTPUT_FILE))
```
